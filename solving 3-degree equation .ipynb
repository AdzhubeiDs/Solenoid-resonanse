$$A=(2\pi I^2 \mu N^2 S)/c$$

```python
#import numpy as np
#данные
mu=1
r=3.5 #см
S=np.pi*r**2
l=int(input('length of a spring -'))
m=int(input('mass -'))
k=int(input('koeff of density -'))
N=int(input('number of coils -'))
c=3*10**9
g=9.81 #м/с^2
```

```python
I=int(input('Current- '))

A= 2*np.pi*I**2*mu*N**2*S/c
print('A=', A)

from sympy import solve, Symbol
x = Symbol('x')
b=2*l-m*g/k
c=l**2-m*g*2*l/k
d=(A-m*g*l**2)/k
print(solve( x ** 3 + b * x ** 2 + c * x + d, x))
```

```python
from numpy.polynomial import Polynomial
x = Polynomial([d,c,b,1])
print(x.roots())
```

```python
a=1
q = (3*a*c - b**2) / (9 * a**2)
r = (9*a*b*c - 27*a**2*d - 2*b**3) / (54*a**3)

print ("q = ",q)
print ("r = ",r)

delta = q**3 + r**2

print ("delta = ",delta, '\n')

# here delta is less than zero so we use the second set of equations from the article:

rho = (-q**3)**0.5

# For x1 the imaginary part is unimportant since it cancels out

theta = np.arccos(r/rho)
s_real = rho**(1./3.) * np.cos( theta/3)
t_real = rho**(1./3.) * np.cos(-theta/3)

# print ("s [real] = ",s_real)
# print ("t [real] = ",t_real)

x1 = s_real + t_real - b / (3. * a)
print ("real solution = ", x1)

print ("should be zero: ",a*x1**3+b*x1**2+c*x1+d)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cubic equation
def cubic_eq(x):
    return x**3 +b*x**2 + c*x - d

# Generate x and y coordinates
x = np.linspace(-60, 60, 10000)
y = cubic_eq(x)

# Plot the cubic equation
plt.plot(x, y)
plt.title('Cubic Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, [0 for i in range(len(x))], c='k')
plt.grid()
plt.show()
```
